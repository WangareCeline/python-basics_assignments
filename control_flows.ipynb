```python
# Practice combining and, or, not.
#password = "MyCoolPassword123"
password = 'K3v!n_Sky77_#Pz'
has_special = "!" in password
has_number = any(char.isdigit() for char in password)

#Length over the 8-character minimum).
#No Forbidden Words: It doesn't contain the word "password."
#Complexity: It uses a mix of uppercase letters, lowercase letters, numbers, and special characters (!, _, #).
# Combining them all
if len(password) >= 8 and not ("password" in password.lower()) and (has_special or has_number):
    print("Strong password!")
else:
    print("Password is too weak or contains forbidden words.")
```

```python
# Build a grading system:
# Input score (0–100)
# Output grade: A (90+), B (80–89), etc.

score = int(input("Enter your score:    "))

if score>=90:
    print(f"Your score of {score} is an A")
elif 89>=score>=80:
    print(f"Your score of {score} is a B")
elif 79>=score>=70:
    print(f"Your score of {score} is a C")
elif 69>=score>=60:
    print(f"Your score of {score} is a D")
else:
    print(f"Your score of {score} is an E")
```
