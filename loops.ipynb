```python
# Use a while loop to print numbers until the user enters stop.
active = True

while active:
    user_input = input("Enter a number: ")
    
    if user_input.lower() == "stop":
        active = False
    else:
        print(f"Number: {user_input}")
```

```python
# Write a guessing game that asks the user to guess a secret number between 1 and 10.
# Give feedback (too high / too low)
# Use a while loop

import random

# Generate a secret number between 1 and 10
secret_number = random.randint(1, 10)

print("I'm thinking of a number between 1 and 10.\nTo end the game type stop")

while True:
    user_input = input("Enter your guess: ")

    # Check if the user wants to quit early
    if user_input.lower() == 'stop':
        print(f"Giving up? The number was {secret_number}.")
        break

    # Convert the input to an integer
    guess = int(user_input)

    # Check the guess
    if guess == secret_number:
        print(f"Correct! {secret_number} was the secret number.")
        break  # Game over, exit the loop
    elif guess < secret_number:
        print("Too low! Try again.")
    else:
        print("Too high! Try again.")
```
