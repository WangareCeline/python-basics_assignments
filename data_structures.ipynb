```python
# Create a program that:
# Takes 5 user inputs and stores them in a list
# Converts the list into a set and prints the unique values

user_input = input('Enter five numbers(separate them with commas)')
input_list = [input for input in user_input.split(',')]
input_set = set(input_list)
print(f"{input_list}\n{input_set}")
```
